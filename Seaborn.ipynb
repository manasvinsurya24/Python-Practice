```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
df = sns.load_dataset("tips")
df.head()
```

```python
df.total_bill
```

```python
sns.distplot(df["total_bill"],kde = False, bins = 100)
```

```python
sns.jointplot(x = 'total_bill', y = 'size', data = df,kind = "hex")
```

```python
sns.pairplot(df, hue = "smoker")
```

```python
sns.rugplot(df["tip"])
```

```python
import numpy as np
sns.barplot(x = "sex" , y = "tip" , data = df , estimator = np.mean)
```

```python
sns.countplot(x = "sex" , data = df )
```

```python
sns.boxplot(x = "day" , y = "tip" , data =df , hue = "sex")
```

```python
sns.violinplot(x = "day" , y = "tip" , data =df)
```

```python
sns.stripplot(x = "day" , y = "tip" , data =df , hue = "sex" , jitter = True , split = True)
```

```python
sns.swarmplot(x = "day" , y = "tip" , data =df )
```

```python
sns.factorplot( x = "day" , y = "total_bill" , data = df , kind = "box")
```
