```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing
from mlxtend.evaluate import bias_variance_decomp
from sklearn.ensemble import ExtraTreesClassifier
```

```python
df = pd.read_csv("telescope_data.csv")
label_encoder = preprocessing.LabelEncoder()
df["class"] = label_encoder.fit_transform(df["class"])
df["class"].unique()
x = df.iloc[:,1:11]
y = df.iloc[:,11]
y
```

```python
imp_features = ExtraTreesClassifier()
imp_features.fit(x,y)
print(imp_features.feature_importances_)
```

```python
feat_importances = pd.Series(imp_features.feature_importances_, index = x.columns)
feat_importances.nlargest(10).plot(kind = "barh")
plt.show()
```

```python
x_new = x.drop(labels = ["fM3Trans","fConc1","fAsym","fConc","fDist"], axis = 1)
x_new.head()
```

```python
df_new = df.drop(labels = ["Unnamed: 0","fM3Long","fConc1","fAsym","fConc","fDist"], axis = 1)
df_new.head()
```

```python
sns.pairplot(df_new, hue = "class")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
from sklearn.model_selection import cross_val_score
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_new,y, random_state = 20)
knnclassifier = KNeighborsClassifier(n_neighbors = 7)
knnclassifier.fit(x_train, y_train)
y_pred = knnclassifier.predict(x_test)
metrics.accuracy_score(y_test, y_pred)
```

```python
y_pred
```

```python
variance = np.var(y_pred)
variance
```

```python
knnclassifier = KNeighborsClassifier(n_neighbors = 7)
print(cross_val_score(knnclassifier, x_new, y, cv = 10, scoring = "accuracy" ).mean())
```

```python
sse = np.mean((np.mean(y_pred) - y_test)**2)
Bias = sse - variance
Bias
```

```python
expected_loss, bias, var = bias_variance_decomp(knnclassifier, x_train.values, y_train.values, x_test.values, y_test.values, loss= 'mse', num_rounds=100, random_seed=53)
```
