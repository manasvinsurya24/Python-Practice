```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(1,2,5)
y = x**3
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel("Numbers")
plt.ylabel("Cubes of N")
plt.title("Title")
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,"r.")
plt.subplot(1,2,2)
plt.plot(y,x,"y.")
plt.show()
```

```python
fig = plt.figure()
axes = fig.add_axes([0.2,0.2,0.9,0.9])
axes.plot(x,y,"r")
axes.set_xlabel("X Label")
axes.set_ylabel("Y Label")
axes.set_title("Title")
axes.plot(y,x,"y")
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.5,0,0.5,0.5])

axes1.plot(x,y,"r")
axes2.plot(y,x,"y")
```

```python
fig,axes = plt.subplots(nrows = 2, ncols = 2,figsize = (10,10))
fig.tight_layout()
axes[0][0].plot(x,x**5)
axes[0][1].plot(x,x**4)
axes[1][0].plot(x,x**3)
axes[1][1].plot(x,x**2)
```

```python
fig.savefig("matplotlib_1.png", dpi = 1000)
```

```python
fig = plt.figure(figsize = (5,5))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig = plt.figure()
axes = fig.add_axes([0.2,0.2,0.9,0.9])
axes.plot(x,x**2, label = "Squared")
axes.plot(x,x**3, label = "Cubed")
axes.set_xlabel("X Label")
axes.set_ylabel("Y Label")
axes.set_title("Title")
axes.legend()
```
