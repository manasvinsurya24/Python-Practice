```python
import pandas as pd
reactions = pd.read_csv("Reactions.csv")
user = pd.read_csv("User.csv")
session = pd.read_csv("Session.csv")
reactiontypes = pd.read_csv("ReactionTypes.csv")
profile = pd.read_csv("Profile.csv")
content = pd.read_csv("Content.csv")
location = pd.read_csv("Location.csv")
```

```python
print(reactions.info())
print(user.info())
print(session.info())
print(reactiontypes.info())
print(profile.info())
print(content.info())
print(location.info())
```

```python
df_user = pd.merge(pd.merge(pd.merge(user,profile,on='User ID' , how = "inner"),session,on="User ID" , how = "inner"),location , on = "User ID", how = "inner")
df_user.drop(["Unnamed: 0_x","Unnamed: 0_y"] , axis =1 , inplace = True)
df_user
```

```python
df_r = pd.merge(reactions,reactiontypes, on ="Type" , how = 'inner')
df_r.drop(["Unnamed: 0_x","Unnamed: 0_y"] , axis = 1, inplace = True)
df_r
```

```python
df_user_r = pd.merge( df_user ,df_r , on = "User ID", how = "inner")
df_user_r
```

```python
df_user_r.groupby("User ID").count()
```

```python
df_user_r.groupby("Content ID").describe().transpose()
```

```python
df_final = pd.merge(df_user_r, content, on = "Content ID", how = "inner")
df_final["Type_media"] = df_final["Type_y"]
df_final["Type_content"] = df_final["Type_x"]
df_final.drop(["Unnamed: 0","Type_y","Type_x"], axis = 1, inplace = True)
df_final
```

```python
df_final.drop(["Datetime","User ID_x","Name","Email","Address","User ID_y"], axis = 1 ,inplace = True)
```
